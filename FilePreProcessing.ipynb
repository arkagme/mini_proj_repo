```python
import pandas as pd
```

```python
df=pd.read_csv('file.csv')
```

```python
df.drop(["id"],axis=1)
```

```python
import pickle

with open('assignments.pkl', 'rb') as f:
    self_dict = pickle.load(f)

with open('shared.pkl', 'rb') as f:
    share_dict = pickle.load(f)

import warnings
warnings.filterwarnings("ignore")
self = []
for i in df.index:
    if df['pc'][i] == self_dict[df['user'][i]]:
        self.append('self')
    elif df['pc'][i] in share_dict:
        self.append('shared')
    else:
        self.append('other')
df['pc'] = self
```

```python
df
```

```python
df["locality"] = "No-info"
for index, row in df.iterrows():
    if row["filename"].startswith("R:"):
        df.loc[index, "locality"] = "External"
    if row["filename"].startswith("C:"):
        df.loc[index, "locality"] = "Local" 
    
        
    
```

```python
df
```

```python
df.info()
```

```python
df['activity'] = df['activity'].apply(lambda x : x.replace("File ",""))
```

```python

df["remove"] = "none"
for i in range(df.shape[0]):
    if df["to_removable_media"][i] == True and (df["activity"][i] == 'Copy'):
        df.loc[i, "remove"] = "To"
    elif df["from_removable_media"][i] == True and (df["activity"][i] == 'Copy'):
        df.loc[i,"remove"] = "From"
    else:
        df.loc[i,"remove"] = " "
```

```python
df['remove']
```

```python
decoy = pd.read_csv('decoy_file.csv',index_col = 'pc')
dfdec = []
for i in range(df.shape[0]):
    if df['user'][i] in decoy.index and decoy.loc[df['user'][i]] == df['filename'][i]:
        dfdec.append('decoy')
    else:
        dfdec.append('not-decoy')
df['decoy'] = dfdec
```

```python
df
```

```python
df['token'] = df["pc"]+"_"+df["locality"]+"_"+df["activity"]+"_"+df["remove"]+'_'+df['decoy']
```
