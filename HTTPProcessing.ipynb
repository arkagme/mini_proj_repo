```python
from urllib.parse import urlparse, parse_qs
import re

def tokenize_url(url):
    # Parse the URL
    parsed_url = urlparse(url)
    
    # Tokenize domain
    domain_tokens = parsed_url.netloc.lower().split('.')
    
    # Tokenize path
    path_tokens = [token.lower() for token in parsed_url.path.split('/') if token]
    
    # Tokenize query parameters
    query_tokens = []
    if parsed_url.query:
        query_params = parse_qs(parsed_url.query)
        query_tokens = [f"{k.lower()}:{v[0].lower()}" for k, v in query_params.items()]
    
    # Combine all tokens
    all_tokens = domain_tokens + path_tokens + query_tokens
    
    # Remove special characters and filter out empty tokens
    cleaned_tokens = [re.sub(r'[^a-z0-9]', '', token) for token in all_tokens]
    cleaned_tokens = [token for token in cleaned_tokens if token]
    
    return cleaned_tokens
```

```python
df['script'] = df['url'].apply(lambda x : x.split(sep = '.')[-1])
```

```python
df = df.drop(['url'],axis=1)
```

Token Sequence = PC + action + flag

PC:  self , other or shared


action: download, upload or visit


flag: sus or safe  script

```python
set(df['script'].values)
```

```python
def sus(x):
    if x in 'doc,pdf,zip,jsp,php'.split(sep=','):
        return 'sus'
    return 'saf'
df['flag'] = df['script'].apply(lambda x : sus(x))
```

```python
import pickle

with open('assignments.pkl', 'rb') as f:
    self_dict = pickle.load(f)
```

```python
with open('shared.pkl', 'rb') as f:
    share_dict = pickle.load(f)
```

```python
import warnings
warnings.filterwarnings("ignore")
self = []
for i in df.index:
    if df['pc'][i] == self_dict[df['user'][i]]:
        self.append('self')
    elif df['pc'][i] in share_dict:
        self.append('shared')
    else:
        self.append('other')
df['pc'] = self
```

```python
df['token'] = df['pc']+'_'+df['action']+'_'+df['flag']
```

```python
df = df.drop(['pc','script','action','flag'],axis=1)
df
```
